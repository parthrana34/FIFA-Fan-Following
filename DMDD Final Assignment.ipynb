## Web Scraping

```python
# Assignment 1 for INFO 6210: Database Management and Database Design
# Prgramming Language Used: Python
# Group: Shaishav Shah and Parth Rana
# Due Date: 30th January 2019 (Wednesday)

# Objective - Extracting top 100 soccer player's basic information and their statistics as per Fifa 17 using web scrapping

# Importing all the required libraries for web scraping
from bs4 import BeautifulSoup
import urllib3 # for opening and reading data through URL
import requests
import pandas as pd
import numpy as np

# Creating a list of pages from which observations are to be extracted
pages = list()
n = 0
while n < 4:
    url='https://www.futwiz.com/en/fifa17/career-mode/players?page=0&order=rating&s=desc&page='
    combine = url + str(n)
    pages.append(combine)
    n = n+1
```

```python
## Code of extracting each football player's basic information.

site = list()
stats = list()
stat= list()
temp = list()
dataset = list()

# Exracting player's basic information from the main page.
for page in pages:
    http = urllib3.PoolManager()
    response = http.request('GET',page)
    soup = BeautifulSoup(response.data,'lxml')
    
    res = soup.find_all('tr', attrs ={'class':'table-row'})
    
    for obs in res:
        name = obs.find('b').text
        team = obs.find('p', attrs = {'class': 'team'}).text.strip()[:-2]
        ovr = obs.find('div').text
        pot = obs.find_all('td')[3].text.strip()
        pos = obs.find_all('td')[4].text.strip()
        grow = obs.find_all('td',attrs={'class':'statCol'})[3].text.strip()
        age = obs.find('td',attrs={'class':'statCol text-center'}).text.strip()
        contract = obs.find_all('td',attrs={'class':'statCol text-center'})[1].text.strip()
        dataset.append((name,team,ovr,pot,pos,grow,age,contract)) #Creating a list with basic information
    
    #Finding urls for extracting statistics for each football player
    for sval in res:
        abc = sval.find(href = True)
        temp.append(abc)

#Creating a list of urls for extracting statistics for each football player
for sval in temp:
    player = 'https://www.futwiz.com'+sval['href']
    site.append(player)
    if player == 'https://www.futwiz.com/en/fifa17/career-mode/player/nemanja-matic/4714': break
            
    
```

```python
## Code of extracting each football player's individual statistics.


for info in site:
    http = urllib3.PoolManager()
    response = http.request('GET',info)
    soup = BeautifulSoup(response.data,'lxml')
    
    stats = soup.find_all('div', attrs = {'class': 'col-2'})

    #Creating a list with all of their individual statistics as per Fifa 17 ratings.
    for obs in stats:
        acc = stats[10].find('div',attrs = {'class': 'individual-stat-bar-stat textcolour accelerationstat'}).text
        sprint = stats[10].find('div',attrs = {'class': 'individual-stat-bar-stat textcolour sprintspeedstat'}).text
        positioning =  stats[11].find('div',attrs = {'class': 'individual-stat-bar-stat textcolour positioningstat'}).text
        fin = stats[11].find('div',attrs = {'class':'individual-stat-bar-stat textcolour finishingstat'}).text
        long_shoot = stats[11].find('div',attrs = {'class':'individual-stat-bar-stat textcolour longshotstat'}).text
        penalty = stats[11].find('div',attrs = {'class':'individual-stat-bar-stat textcolour penaltiesstat'}).text
        power_shot = stats[11].find('div',attrs = {'class':'individual-stat-bar-stat textcolour shotpowerstat'}).text
        volley = stats[11].find('div',attrs = {'class':'individual-stat-bar-stat textcolour volleysstat'}).text
        cross = stats[12].find('div',attrs = {'class':'individual-stat-bar-stat textcolour crossingstat'}).text
        curve = stats[12].find('div',attrs = {'class':'individual-stat-bar-stat textcolour curvestat'}).text
        fk = stats[12].find('div',attrs = {'class':'individual-stat-bar-stat textcolour fkaccstat'}).text
        long_pass = stats[12].find('div',attrs = {'class':'individual-stat-bar-stat textcolour longpassstat'}).text
        short_pass = stats[12].find('div',attrs = {'class':'individual-stat-bar-stat textcolour shortpassstat'}).text
        vision = stats[12].find('div',attrs = {'class':'individual-stat-bar-stat textcolour visionstat'}).text
        agility = stats[13].find('div',attrs = {'class':'individual-stat-bar-stat textcolour agilitystat'}).text
        bal = stats[13].find('div',attrs = {'class':'individual-stat-bar-stat textcolour balancestat'}).text
        ball_ctr = stats[13].find('div',attrs = {'class':'individual-stat-bar-stat textcolour ballcontrolstat'}).text
        drb = stats[13].find('div',attrs = {'class':'individual-stat-bar-stat textcolour dribblingstat'}).text
        react = stats[13].find('div',attrs = {'class':'individual-stat-bar-stat textcolour reactionsstat'}).text
        head = stats[14].find('div',attrs = {'class':'individual-stat-bar-stat textcolour headingaccstat'}).text
        intercept = stats[14].find('div',attrs = {'class':'individual-stat-bar-stat textcolour tactawarestat'}).text
        mark = stats[14].find('div',attrs = {'class':'individual-stat-bar-stat textcolour markingstat'}).text
        std_tackle = stats[14].find('div',attrs = {'class':'individual-stat-bar-stat textcolour standingtacklestat'}).text
        slide = stats[14].find('div',attrs = {'class':'individual-stat-bar-stat textcolour slidetacklestat'}).text
        agg = stats[15].find('div',attrs = {'class':'individual-stat-bar-stat textcolour aggressionstat'}).text
        jump = stats[15].find('div',attrs = {'class':'individual-stat-bar-stat textcolour jumpingstat'}).text
        stamina = stats[15].find('div',attrs = {'class':'individual-stat-bar-stat textcolour staminastat'}).text
        strength = stats[15].find('div',attrs = {'class':'individual-stat-bar-stat textcolour strengthstat'}).text

    #List of individual attributes of all the football players.
    stat.append((acc,sprint,positioning,fin,long_shoot,penalty,power_shot,volley,cross,curve,fk,long_pass,short_pass,vision,
                 agility,bal,ball_ctr,drb,react,head,intercept,mark,std_tackle,slide,agg,jump,stamina,strength))
    
    
```

```python
## Creating individual dataframes for basic info and attributes associated to each player. 

basic = pd.DataFrame(dataset, columns = ['Name1','Team','Overall','Potential','Position','Growth','Age','Contract'])

statistics = pd.DataFrame(stat, columns = ['Acceration_17','Sprint_Speed_17','Positiong_17','Finishing_17','Long_Shoot_17',
                                           'Penalty_17','Power_Shot_17','Volley_17','Crossing_17','Curve_17','FK_Accuracy_17','Long_Pass_17'
                                           ,'Short_Pass_17','Vision_17','Agility_17','Balance_17','Ball_Control_17','Dribble_17','Reaction_17','Heading_17'
                                           ,'Interceptions_17','Marking_17','Standing_Tacke_17','Sliding_Tackle_17','Aggression_17','Jumping_17','Stamina_17'
                                           ,'Strength_17'])
```

```python
## Creating final dataframe for will all the basic info and attributes associated to each player. 

Fifa_17 = pd.concat([basic,statistics],axis =1)

## Creating unique ID for final dataframe
Fifa_17.reset_index(inplace =True)
Fifa_17 = Fifa_17.rename( columns = {'index': 'Unique_ID'})


## Final Dataframe using web Scraping (first Five row)
Fifa_17.head()
```

## Data Extract using Web API

```python
## Objective - Extracting top 100 soccer player's followers, desciption and twitter user id as per Fifa 17 using Twitter API

# Creating a list of official user id used by top 100 soccer players as per Fifa 17 from google using web scraping

# list of urls for soccer players
twitter = list()
for text in Fifa_17['Name1']:
    url = 'http://www.google.com/search?q='+text+' official twitter account'
    twitter.append(url)
```

```python
# Extracting individual player's google page which contains their user id.
ser_name = list()
for data in twitter:
    handle = requests.get(data).text
    for res in handle.split():
        if '(@' in res:
            user = res
            break
    user_name.append(user[2:-1])
```

```python
# Cleaning the data and extracting the user id from google page.
final = list()
for a in user_name:
    if a.startswith('<'):
        final.append(a[3:-4])
    elif a.startswith('_'):
        final.append(a[4:-4])
    else:

        
        final.append(a)
```

```python
## Gathering data like followers and twitter page description using twitter API

#Code for authenticating the user to access twitter as a developer

#Importing tweepy library for gathering required data using API
import tweepy

# Important cridentials for accessing twitter acount as a developer
consumer_key = "PUzlj7GqfS5AyWrXAIQfLk9uT" # Use your own key. To get a key https://apps.twitter.com/
consumer_secret = "WTnbRXIeTsDiKfPKm6m3N7XqrgmqU7wwaQlCu0dKAkQAa6lkxr"
access_token = '842436844629319680-vm8u4c1lCdEO5GnDjPolUwK63ON2qUQ'
access_token_secret = '51wxVXxArkVMcLgThIwNP3U8aZj3nrAIFixkec4Qbqtsv'

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)

api = tweepy.API(auth)
```

```python
## Code for finding followers and page description for individual players.

dataset = list()
for a in final:
    user = api.get_user(a)
    fc = user.followers_count
    sn = user.screen_name
    des = user.description
    dataset.append((sn,fc,des))
```

```python
## Code for creating final dataframe of the data extracted by API

Twitter = pd.DataFrame(dataset, columns = ['Screen_Name','Followers_Count','Twitter_Page_Description'])
Fifa_Twitter = pd.concat([Fifa_17['Name1'],Twitter],axis =1)
Fifa_Twitter = Fifa_Twitter.rename( columns = {'Name1': 'Name2'})

## Creating unique ID for final dataframe
Fifa_Twitter.reset_index(inplace =True)
Fifa_Twitter = Fifa_Twitter.rename( columns = {'index': 'Unique_ID'})

## Final DataFrame using Twitter API (first Five row)
Fifa_Twitter.head()
```

## Data Extraction from CSV file

```python
## Objective - Extracting top 100 soccer player's basic infor and statistics as per Fifa 19.
## Dataset Scource: Kaggle


# Importing pandas library for importinf CSV file from system
import pandas as pd
data = pd.read_csv('Fifa19.csv') #Reading CSV file
file = data[0:1000] #taking top 1000 data points out of approx 18000

# Cleaning data by removing unwanted attributes
file = file.drop(['Nationality','ID','Unnamed: 0','Photo','Flag','Club Logo','Value','Wage','Special','Preferred Foot','International Reputation','Weak Foot','Skill Moves','Work Rate','Body Type','Real Face','Jersey Number','Joined','Loaned From','Contract Valid Until','Height','Weight','LS','ST','RS','LW','LF','CF','RF','RW','LAM','CAM','RAM','LM','LCM','CM','RCM','RM','LWB','LDM','CDM','RDM','RWB','LB','LCB','CB','RCB','RB','GKDiving','GKHandling','GKKicking','GKPositioning','GKReflexes','Release Clause'],axis=1)
file = file.rename( columns = {'Name': 'Name3'})
```

```python
## Code for sorting Fifa 19 data according to Fifa 17 data set.

nlist = list()
for player in Fifa_17['Name1']:
    n = 0
    for each in file['Name3']:
        if player.split()[-1] == each.split()[-1]:
            nlist.append((n))
            break
        elif n == 999:
            nlist.append('-1')
        n = n+1
        
Fifa_19 = file.iloc[nlist]

##Structuring required data set from final dataframe
Fifa_19.reset_index(inplace = True)
Fifa_19.reset_index(inplace = True)
Fifa_19 = Fifa_19.drop('index', axis =1)
Fifa_19 = Fifa_19.rename( columns = {'level_0': 'Unique_ID'})
Fifa_19 = Fifa_19.rename( columns = {'Name': 'Name3'})

#Final Dataframe using dataset from Kaggle (first Five row)
Fifa_19.head()
```
